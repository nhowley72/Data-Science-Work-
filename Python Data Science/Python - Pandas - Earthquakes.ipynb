# Task 2.4. There are 10 outliers in the data set you have been provided. In particular, for these ten entries, the magnitude is a negative value. Find these 10 outliers and output them in a dataframe containing the time, place, magnitude, latitude and longitude. Then change the sign of these ten entries to positive.

```python
# Lambda function for absolute value
AbsoluteValue = lambda x: abs(x)
L = []  # list of outliers

# Load the data
df = pd.read_csv("earthquakes.csv")
dfOutliers = df[['time', 'place', 'mag', 'latitude', 'longitude']]  # columns we care about from df
dfOutliersToPrint = pd.DataFrame(columns=['time', 'place', 'mag', 'latitude', 'longitude'])  # DataFrame to show answer later

# Identify outliers
for i in range(len(dfOutliers.index)):  # for every index value
    if dfOutliers.mag.values[i] < 0:  # check if mag is negative
        L.append(i)  # if it is negative, add to the list of outlier indices

# Correct outliers by applying the absolute value
dfOutliers['mag'] = dfOutliers['mag'].apply(AbsoluteValue)  # amend the outlier values

# Append outlier rows to the answer DataFrame
for i in L:
    dfOutliersToPrint = pd.concat([dfOutliersToPrint, dfOutliers.loc[[i]]], ignore_index=True)

# Sort the values for a nicer look and print the result
print(dfOutliersToPrint.sort_index(ascending=True))
```

##  3.1 Create a dataframe containing the following descriptive statistics as series:
• Mean earthquake magnitude,
• Standard deviation of the magnitude,
• Median earthquake magnitude,
• Number of earthquakes in the dataset.
• Date of the first earthquake.
• Date of the last earthquake.
for all of the following areas: Alaska, Canada, Washington, Oregon, California and Mexico

```python
import pandas as pd

def CollectStats(df, Country, dfStats): 
    # Method for collecting statistics with arbitrary DataFrame and Country name
    MagMean = df['mag'].mean()
    MagStd = df['mag'].std()
    MagMed = df['mag'].median()
    NumberOfEQS = len(df.index)
    DateFirst = df['time'].min()
    DateLast = df['time'].max()
    
    # Create a new DataFrame row with statistics
    Row = pd.DataFrame({
        'Mean Earthquake Mag': [MagMean], 
        'Std Dev Of Mag': [MagStd],
        'Median Earthquake Mag': [MagMed], 
        'Number of Earthquakes': [NumberOfEQS],
        'Date Of First Earthquake': [DateFirst],
        'Date Of Last Earthquake': [DateLast],
        'Country Name': [Country]
    })
    
    # Append the new row to dfStats using pd.concat()
    dfStats = pd.concat([dfStats, Row], ignore_index=True)
    
    print(dfStats)  # Display the updated DataFrame for reference
    return dfStats  # Return the updated DataFrame

# Fill NaNs in the 'place' column
df['place'] = df['place'].fillna("")

# Create individual DataFrames for each place of interest
dfAlaska = df.loc[df['place'].str.contains('Alaska', case=False)]
dfCanada = df.loc[df['place'].str.contains('Canada', case=False)]
dfWashington = df.loc[df['place'].str.contains('Washington', case=False)]
dfOregon = df.loc[df['place'].str.contains('Oregon', case=False)]
dfCalifornia = df.loc[df['place'].str.contains('California', case=False)]
dfMexico = df.loc[df['place'].str.contains('Mexico', case=False)]

# List of DataFrames and areas for easier iteration
ListDfs = [dfAlaska, dfCanada, dfWashington, dfOregon, dfCalifornia, dfMexico]
ListAreas = ['Alaska', 'Canada', 'Washington', 'Oregon', 'California', 'Mexico']

# Define columns for the statistics DataFrame
ListCols = [
    "Mean Earthquake Mag", "Std Dev Of Mag", "Median Earthquake Mag",
    "Number of Earthquakes", "Date Of First Earthquake", 
    "Date Of Last Earthquake", "Country Name"
]
dfStats = pd.DataFrame(columns=ListCols)  # Create the statistics DataFrame

# Populate the statistics DataFrame for each area
for i in range(len(ListDfs)):
    dfStats = CollectStats(ListDfs[i], ListAreas[i], dfStats)

# Final display of the collected statistics
print(dfStats)
```

## 3.2 For each of the data sets in Task 3.1, determine the mean earthquake frequency and the average period that elapses between successive earthquakes (you do not need to account for aftershocks). Calculate the mean frequency in earthquakes/day. Calculate the mean period by calculating all the inter-earthquake intervals in days and divide by the number of intervals. Present your results in a dataframe.


```python
import pandas as pd
from datetime import datetime

# Columns for the frequency DataFrame
ListCols2 = ["Mean Earthquake Frequency", "Average Period Between Successive Earthquakes", "CountryName"]
dfFreq = pd.DataFrame(columns=ListCols2)  # New DataFrame to store frequency data

# Function to calculate mean earthquake frequency and period between earthquakes
def meanEqFreq(dfCountry, dfFreq, Country): 
    # Convert max and min time to datetime for calculation
    dateMax = pd.to_datetime(dfCountry['time'].max())
    dateMin = pd.to_datetime(dfCountry['time'].min())
    
    # Calculate the interval in days
    dateInterval = dateMax - dateMin
    days = dateInterval.days
    
    # Calculate mean earthquake frequency and average period
    MeanEQFREQ = len(dfCountry.index) / days  # Frequency = number of events / time interval in days
    meanPeriod = 1 / MeanEQFREQ  # Period = 1 / frequency
    
    # Create a new row with the calculated values
    Row = pd.DataFrame({
        "Mean Earthquake Frequency": [MeanEQFREQ],
        "Average Period Between Successive Earthquakes": [meanPeriod],
        "CountryName": [Country]
    })
    
    # Append the row to dfFreq using pd.concat
    dfFreq = pd.concat([dfFreq, Row], ignore_index=True)
    return dfFreq

# Apply meanEqFreq function to each DataFrame in ListDfs
for i in range(len(ListDfs)):
    dfFreq = meanEqFreq(ListDfs[i], dfFreq, ListAreas[i])

# Display the final DataFrame with frequency statistics
print(dfFreq)
```

```python
import matplotlib.pyplot as plt
import scipy.stats
import numpy as np

# Loop through the DataFrames in ListDfs
for i in range(len(ListDfs)):  
    # Check if 'mag' column exists in the DataFrame
    if 'mag' not in ListDfs[i].columns:
        print(f"Warning: The DataFrame for {ListAreas[i]} does not contain a 'mag' column.")
        continue  # Skip this DataFrame if it lacks the 'mag' column
    
    fig, ax = plt.subplots()
    
    # Set the number of bins
    bins = 60 if i == 6 else 20
    
    # Extract magnitude data, skipping missing values
    data = ListDfs[i]['mag'].dropna()
    
    # Plot the histogram with density normalization
    _, bin_edges, _ = ax.hist(data, bins=bins, density=True, alpha=0.5, color='skyblue', label='Histogram')
    
    # Calculate mean and standard deviation
    mu = data.mean()
    sigma = data.std()
    
    # Generate a best-fit line for the normal distribution
    best_fit_line = scipy.stats.norm.pdf(bin_edges, mu, sigma)
    
    # Plot the fitted normal distribution
    ax.plot(bin_edges, best_fit_line, color='red', label='Fitted distribution')
    
    # Assign titles and labels
    ax.set_title(f'Magnitude of Earthquakes in {ListAreas[i]}')
    ax.set_xlabel('Magnitude')
    ax.set_ylabel('Density')
    ax.legend()

    # Show the plot
    plt.show()
```
